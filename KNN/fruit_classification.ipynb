```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n = 100
apple_width = np.random.normal(6,1,n)
apple_length = np.random.normal(6,1,n)
banana_width = np.random.normal(4,1,n)
banana_length = np.random.normal(8,1,n)
watermelon_width = np.random.normal(10,1,n)
watermelon_length = np.random.normal(10,1,n)
```

```python
fig = plt.figure(figsize=(8, 6))
fig.patch.set_facecolor('#2b2b2b')  
ax = plt.gca()
ax.set_facecolor('#2b2b2b')  

plt.scatter(apple_width, apple_length, c='crimson', marker='o', label='Apple')
plt.scatter(banana_width, banana_length, c='royalblue', marker='o', label='Banana')
plt.scatter(watermelon_width, watermelon_length, c='forestgreen', marker='o', label='Watermelon')



plt.title('Apple, Banana, Watermelon', fontsize=14, color='#e0e0e0')
plt.xlabel('Width', fontsize=12, color='#e0e0e0')
plt.ylabel('Length', fontsize=12, color='#e0e0e0')


plt.xticks(fontsize=10, color='#e0e0e0')
plt.yticks(fontsize=10, color='#e0e0e0')


ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#e0e0e0')
ax.spines['bottom'].set_color('#e0e0e0')


plt.grid(color='#555555', linestyle='--', linewidth=0.5, alpha=0.6)


plt.legend(loc='upper left', fontsize=10, facecolor='#2b2b2b', edgecolor='#555555', framealpha=0.8)


plt.tight_layout()


plt.show()
```

```python
width = np.concatenate((apple_width,banana_width,watermelon_width))
length = np.concatenate((apple_length,banana_length,watermelon_length))
x_train = np.array([width, length]).T
x_train.shape
```

```python
y_apple = np.zeros(n, dtype='int')
y_banana= np.ones(n, dtype='int')
y_watermelon = np.ones(n, dtype='int')*2
y_train = np.concatenate((y_apple, y_banana, y_watermelon))
y_train
```

```python
def euclidean_distance(x1,x2):
    return np.sqrt(np.sum(x1-x2)**2)
```

```python
new_fruit = np.array([8,2])

distance = []
for x in x_train:
    d = euclidean_distance(new_fruit, x)
    distance.append(d)

distance
```

```python
k = 5
```

```python
nearest_neighbors = np.argsort(distance)[0:k]
nearest_neighbors
```

```python
y_train[nearest_neighbors]
```

```python
result =np.bincount(y_train[nearest_neighbors])
result
```

```python
if np.argmax(result)==0:
    print('New Fruit Is: Apple')
elif np.argmax(result)==1:
    print('New Fruit Is: Banana')
else:
    print('New Fruit Is: Apple')
```
