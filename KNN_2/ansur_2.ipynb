```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
female_data = pd.read_csv("ANSUR II FEMALE Public.csv")
female_data.columns = female_data.columns.str.lower().str.strip().str.replace(' ', '_')
female_data.head()
```

```python
male_data =pd.read_csv("ANSUR II MALE Public.csv", encoding= "latin1")
male_data.columns = male_data.columns.str.lower().str.strip().str.replace(' ', '_')
male_data.head()
```

```python
data = pd.concat([female_data,male_data])
data
```

```python
data.isnull().sum()
```

```python
data["weightkg"]= data["weightkg"]/10
data["stature"] = data["stature"]/10
data["biacromialbreadth"] = data["biacromialbreadth"]/10
data["bideltoidbreadth"] = data["bideltoidbreadth"]/10


data["gender"] = data["gender"].replace(["Female", "Male"], [0,1]) 
men_data = data[data['gender'] == 0]
print(men_data.head())
```

```python
women_data = data[data['gender'] == 0]  
men_data = data[data['gender'] == 1]    


fig = plt.figure(figsize=(8, 6))
fig.patch.set_facecolor('#2b2b2b')  
ax = plt.gca()
ax.set_facecolor('#2b2b2b')

plt.scatter(women_data['stature'], women_data['weightkg'], c='crimson', marker='o', label='Women', alpha=1.0)  
plt.scatter(men_data['stature'], men_data['weightkg'], c='royalblue', marker='o', label='Men', alpha=0.1)  

plt.title('Height vs Weight for Women and Men', fontsize=14, color='#e0e0e0')
plt.xlabel('Height (meters)', fontsize=12, color='#e0e0e0')
plt.ylabel('Weight (kg)', fontsize=12, color='#e0e0e0')

plt.xticks(fontsize=10, color='#e0e0e0')
plt.yticks(fontsize=10, color='#e0e0e0')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#e0e0e0')
ax.spines['bottom'].set_color('#e0e0e0')

plt.grid(color='#555555', linestyle='--', linewidth=0.5, alpha=0.6)

plt.legend(loc='upper left', fontsize=10, facecolor='#2b2b2b', edgecolor='#555555', framealpha=0.8)

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X = data[['biacromialbreadth', 'bideltoidbreadth', 'weightkg', 'stature']]
y = data['gender']


X_train, X_test, y_train, y_test = train_test_split(X.to_numpy(), y.to_numpy(), test_size=0.2, random_state=42)

print(f"Training features shape: {X_train.shape}, Training labels shape: {y_train.shape}")
print(f"Testing features shape: {X_test.shape}, Testing labels shape: {y_test.shape}")
```

```python
from knn import KNN

model = KNN(k=5)
model.fit(X_train, y_train)
accuracy = model.evaluate(X_test, y_test)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
from knn import KNN

model = KNN(k=3)
model.fit(X_train, y_train)
accuracy = model.evaluate(X_test, y_test)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
from knn import KNN

model = KNN(k=7)
model.fit(X_train, y_train)
accuracy = model.evaluate(X_test, y_test)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
conff_matrix = np.zeros((2, 2), dtype=int)


for i in range(len(y_test)):  
    if y_test[i] == 0 and y_pred[i] == 0:
        conff_matrix[0][0] += 1  # (TN)
    elif y_test[i] == 0 and y_pred[i] == 1:
        conff_matrix[0][1] += 1  # (FP)
    elif y_test[i] == 1 and y_pred[i] == 0:
        conff_matrix[1][0] += 1  # (FN)
    elif y_test[i] == 1 and y_pred[i] == 1:
        conff_matrix[1][1] += 1  # (TP)


print("Confusion Matrix:")
print(conff_matrix)
```

```python

from sklearn import metrics



cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=conff_matrix, display_labels=[0, 1])

fig = plt.figure(figsize=(8, 6))
fig.patch.set_facecolor('#2b2b2b') 
ax = plt.gca()
ax.set_facecolor('#2b2b2b')  

cm_display.plot(ax=ax, cmap='plasma') 
plt.title('Confusion Matrix', fontsize=14, color='#e0e0e0')  
plt.xlabel('Predicted Labels', fontsize=12, color='#e0e0e0')
plt.ylabel('Actual Labels', fontsize=12, color='#e0e0e0')


plt.xticks(fontsize=10, color='#e0e0e0')
plt.yticks(fontsize=10, color='#e0e0e0')


ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#e0e0e0')
ax.spines['bottom'].set_color('#e0e0e0')
plt.grid(color='#555555', linestyle='--', linewidth=0.5, alpha=0.6)


plt.tight_layout()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

knn_model = KNeighborsClassifier(n_neighbors=5)  


knn_model.fit(X_train, y_train)


y_pred = knn_model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print(f"Scikit-learn KNN Model Accuracy: {accuracy * 100:.2f}%")


conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```
